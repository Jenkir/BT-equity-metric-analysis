```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_2016 = "MDworkforce_2016.csv"
data_2017 = "MDworkforce_2017.csv"
data_2018 = "MDworkforce_2018.csv"
data_2019 = "MDworkforce_2019.csv"
data_2021 = "MDworkforce_2021.csv"
```

```python
df_2016 = pd.read_csv(data_2016)
df_2016
```

```python
df_2017 = pd.read_csv(data_2017)
df_2017
```

```python
dfs1 = pd.concat([df_2016, df_2017], axis=0, sort=False)
dfs1
```

```python
df_2018 = pd.read_csv(data_2018)
df_2018
```

```python
df_2019 = pd.read_csv(data_2019)
df_2019
```

```python
dfs2 = pd.concat([df_2018, df_2019], axis=0, sort=False)
dfs2
```

```python
df_2021 = pd.read_csv(data_2021)
df_2021
```

```python
workforce_df1 = pd.concat([dfs1, dfs2], axis=0, sort=False)
workforce_df1
```

```python
workforce_df = pd.concat([workforce_df1, df_2021], axis=0, sort=False)
workforce_df
```

```python
workforce_df[["Location", "Year"]] = workforce_df.Year.str.split(",", expand=True)
workforce_df
```

```python
# Export file as a csv file; will then import into "Build_workforce2016-2021.ipynb"
workforce_df.to_csv("MD_workforce2016-2021.csv", index=False, header=True)
```
