```python
# import csv file & save the data from the csv in variable

unemployment = "Tableau_csvs/unemployment_Balt_MD.csv"
```

```python
df_unemployment = pd.read_csv(unemployment)
df_unemployment
```

```python
df_unemployment.dtypes
```

```python
df_unemployment["Year"] = df_unemployment["Year"].astype(str)
```

```python
df_unemployment = df_unemployment.drop(["White alone", "Black or African American alone"], axis=1)
```
