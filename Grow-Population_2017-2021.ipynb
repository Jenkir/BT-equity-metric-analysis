```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_2017 = "ACS_population_2017.csv"
data_2018 = "ACS_population_2018.csv"
data_2019 = "ACS_population_2019.csv"
data_2021 = "ACS_population_2021.csv"
```

```python
df_2017 = pd.read_csv(data_2017)
df_2017
```

```python
df_2018 = pd.read_csv(data_2018)
df_2018
```

```python
dfs1 = pd.concat([df_2017, df_2018], axis=0, sort=False)
dfs1
```

```python
df_2019 = pd.read_csv(data_2019)
df_2019
```

```python
df_2021 = pd.read_csv(data_2021)
df_2021
```

```python
dfs2 = pd.concat([df_2019, df_2021], axis=0, sort=False)
dfs2
```

```python
pop_df = pd.concat([dfs1, dfs2], axis=0, sort=False)
pop_df
```

```python
pop_df[["Location", "Year"]] = pop_df.Year.str.split(",", expand=True)
pop_df
```

```python
pop_df = pop_df.drop(['Location'], axis=1)
pop_df
```

```python
# Export file as an Excel file
pop_df.to_excel("ACS_population_Tableau.xlsx")
```
