```python
## Build an equitable economy:
### Baltimore City median HH income, Black householder median HH income, white householder median HH income
## 2016-2021
```

```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_2016 = "balt_income_2016.csv"
data_2017 = "balt_income_2017.csv"
data_2018 = "balt_income_2018.csv"
data_2019 = "balt_income_2019.csv"
data_2021 = "balt_income_2021.csv"
```

```python
df_2016 = pd.read_csv(data_2016)
df_2016
```

```python
df_2017 = pd.read_csv(data_2017)
df_2017
```

```python
dfs1 = pd.concat([df_2016, df_2017], axis=0, sort=False)
dfs1
```

```python
df_2018 = pd.read_csv(data_2018)
df_2018
```

```python
df_2019 = pd.read_csv(data_2019)
df_2019
```

```python
dfs2 = pd.concat([df_2018, df_2019], axis=0, sort=False)
dfs2
```

```python
df_2021 = pd.read_csv(data_2021)
df_2021
```

```python
income_df1 = pd.concat([dfs1, dfs2], axis=0, sort=False)
income_df1
```

```python
income_df = pd.concat([income_df1, df_2021], axis=0, sort=False)
income_df
```

```python
# import csv file of Maryland demographics
# first save the data from the csv in variable, "MD" 

MD = "MD_income_2016-2021.csv"
```

```python
# Create dataframe by reading the variable data from the code block above
MD_df = pd.read_csv(MD)
MD_df
```

```python
# combine the dataframes on the common column "Year"
Balt_MD_income = pd.concat([income_df, MD_df], axis=0)
Balt_MD_income
```

```python
# Export file as an Excel file, to be imported into Tableau
Balt_MD_income.to_excel("Build_economy_income_Tableau.xlsx", index=False)
```
