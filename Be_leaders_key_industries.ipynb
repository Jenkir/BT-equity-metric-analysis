```python
# import csv of data downloaded from the Census QWI website (cleaned in Excel)
data_industries = "raw_data_census/QWI_industries_2017-2022_for_jupyter_import.csv"
```

```python
df = pd.read_csv(data_industries)
df.head(10)
```

```python
df.dtypes
```

```python
df = df.astype({'Employees': 'int', 'Industry': 'category'})
```

```python
# consolidate NAICs code industries into single industry category by using .replace; this will replace
# the name of the QWI names of the industries into names used in BT
df["Industry"] = df["Industry"].str.replace("Support Activities for Water Transportation", "Port and logistics")
df["Industry"] = df["Industry"].str.replace("Port and Harbor Operations", "Port and logistics")
df["Industry"] = df["Industry"].str.replace("Media Streaming Distribution Services, Social Networks, and Other Media Networks and Content Providers", "Digital services")
df["Industry"] = df["Industry"].str.replace("Computing Infrastructure Providers, Data Processing, Web Hosting, and Related Services", "Digital services")
df["Industry"] = df["Industry"].str.replace("Computer Systems Design and Related Services", "Digital services")
df["Industry"] = df["Industry"].str.replace("Performing Arts Companies", "Creative")
df["Industry"] = df["Industry"].str.replace("Independent Artists, Writers, and Performers", "Creative")
df["Industry"] = df["Industry"].str.replace("Warehousing and Storage", "Port and logistics")
df["Industry"] = df["Industry"].str.replace("Management, Scientific, and Technical Consulting Services", "Life sciences/ medical devices")
df["Industry"] = df["Industry"].str.replace("Other Professional, Scientific, and Technical Services", "Life sciences/ medical devices")
df["Industry"] = df["Industry"].str.replace("Scientific Research and Development Services", "Life sciences/ medical devices")
df.tail(40)
```

```python
# export as excel file, to be imported into Tableau
df.to_excel("Be_leaders_key_industries.xlsx", index=False)
```
