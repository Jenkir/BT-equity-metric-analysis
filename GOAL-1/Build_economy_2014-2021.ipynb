```python
#Store the chart into a variable (commonly "fig" and display the dictionary using fig.to_dict or 
# fig['data'] to see the elements data or fig['layout'] to review the design of the plot)

# Every Plotly graph is a dictionary
fig = px.line(Goal1_income,
        x='Year',
        y='Median household income',
        color="Location",
        text='Median household income',
        hover_data= {'Location': True,
                     'Population': True},
        title="Median household income")
fig.update_traces(texttemplate='%{text:.2s}', textposition='top center')
fig.show()
```

```python
# Add $ and comma to y axis
fig.update_layout(yaxis_tickprefix = '$', yaxis_tickformat = ',.')
```
