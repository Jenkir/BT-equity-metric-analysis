```python
## Build a stronger workforce:
### Baltimore City & MD
## 2016-2021
```

```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_2016 = "workforce_2016.csv"
data_2017 = "workforce_2017.csv"
data_2018 = "workforce_2018.csv"
data_2019 = "workforce_2019.csv"
data_2021 = "workforce_2021.csv"
```

```python
df_2016 = pd.read_csv(data_2016)
df_2016
```

```python
df_2017 = pd.read_csv(data_2017)
df_2017
```

```python
dfs1 = pd.concat([df_2016, df_2017], axis=0, sort=False)
dfs1
```

```python
df_2018 = pd.read_csv(data_2018)
df_2018
```

```python
df_2019 = pd.read_csv(data_2019)
df_2019
```

```python
dfs2 = pd.concat([df_2018, df_2019], axis=0, sort=False)
dfs2
```

```python
df_2021 = pd.read_csv(data_2021)
df_2021
```

```python
workforce_df1 = pd.concat([dfs1, dfs2], axis=0, sort=False)
workforce_df1
```

```python
workforce_df = pd.concat([workforce_df1, df_2021], axis=0, sort=False)
workforce_df
```

```python
workforce_df[["Location", "Year"]] = workforce_df.Year.str.split(",", expand=True)
workforce_df
```

```python
MD_data = "MD_workforce2016-2021.csv"
```

```python
MD_df = pd.read_csv(MD_data)
MD_df
```

```python
# combine Baltimore df with MD df
workforce_dfs = pd.concat([workforce_df, MD_df], axis=0, sort=False)
workforce_dfs
```

```python
workforce_dfs.dtypes
```

```python
# change data types of 3 columns to integer instead of float (easier to work with in Tableau)
convert_dict = {"Percent graduated high school": int,              
                "Percent earned Associate's degree": int,
                "Percent earned Bachelor's degree":int 
               }
workforce_dfs = workforce_dfs.astype(convert_dict)
workforce_dfs.dtypes
```
