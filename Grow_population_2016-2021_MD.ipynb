```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Will use the below library when we format the axes on our plots
from matplotlib.ticker import FormatStrFormatter
from matplotlib import ticker
```

```python
data_2016 = "MD_population_2016.csv"
data_2017 = "MD_population_2017.csv"
data_2018 = "MD_population_2018.csv"
data_2019 = "MD_population_2019.csv"
data_2021 = "MD_population_2021.csv"
```

```python
df_2016 = pd.read_csv(data_2016)
df_2016
```

```python
df_2017 = pd.read_csv(data_2017)
df_2017
```

```python
dfs1 = pd.concat([df_2016, df_2017], axis=0, sort=False)
dfs1
```

```python
df_2018 = pd.read_csv(data_2018)
df_2018
```

```python
df_2019 = pd.read_csv(data_2019)
df_2019
```

```python
dfs2 = pd.concat([df_2018, df_2019], axis=0, sort=False)
dfs2
```

```python
df_2021 = pd.read_csv(data_2021)
df_2021
```

```python
pop_df1 = pd.concat([dfs1, dfs2], axis=0, sort=False)
pop_df1
```

```python
pop_df = pd.concat([pop_df1, df_2021], axis=0, sort=False)
pop_df
```

```python
pop_df[["Location", "Year"]] = pop_df.Year.str.split(",", expand=True)
pop_df
```
